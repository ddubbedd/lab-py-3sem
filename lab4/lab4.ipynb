```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def changeIndex(df: pd.DataFrame) -> pd.DataFrame:
    """updating indexes DataFrame"""
    df.index = [i for i in range(0, int(df.shape[0]))]
    return df
```

```python

path: str=os.path.join("/Users/aleksandragorbuncova/PycharmProjects/lab-py-3sem/lab3/datasetAll/dataset.csv")

df = pd.read_csv(path)
print(df)
```

```python
df.columns = ['data', 'temp', 'pressure', 'wind']  # 2
print(df)
```

```python
#3
df.replace(" ", float("NaN"), inplace=True)
df = df.dropna()
print(df)
```

```python
df = changeIndex(df)
print(df)
```

```python
df['temp'] = df['temp'].astype(int)  # 4
df = df.assign(temp_f=(df.temp * 9 / 5) + 32)
print(df)
```

```python
stats = df[['temp', "temp_f"]].describe()  # 5
print(stats)
```

```python
def task6(df: pd.DataFrame, v: float) -> pd.DataFrame:
    """sorting DataFrame by value"""
    return df.loc[df['temp'] >= v]
```

```python
v = 15                  # 6
df_t6 = task6(df, v)
df_t6 = changeIndex(df_t6)
print(df_t6)
```

```python
def task7(df: pd.DataFrame, data_1: str, data_2: str) -> pd.DataFrame:
    """sorting DataFrame by data1 and data2"""
    df= df.loc[(df['data'] >= data_1) & (df['data'] <= data_2)]
    return df
```

```python
data1 = "2011-01-10"              # 7
data2 = "2012-01-10"
df_t7 = task7(df, data1, data2)
df_t7 = changeIndex(df_t7)
print(df_t7)
```

```python
figure = plt.figure(figsize=(10, 5))
ax1=figure.add_subplot(211)
ax1.plot(df['temp'], color="purple")
plt.legend(["Temp in day by Celsius"])
plt.xlabel("Day")
plt.ylabel("Temp")

ax2 = figure.add_subplot(212)
ax2.plot(df['temp_f'], color="yellow")
plt.legend(["Temp in day by Fahrenheit"])
plt.xlabel("Day")
plt.ylabel("Temp")
plt.show()
```

```python
def custom(df: pd.DataFrame, month: int, year: int) -> pd.DataFrame:
    """one month schedule"""
    if (month < 10): month = "0" + str(month)
    data1 = str(str(year) + "-" + month + "-" + "01")
    data2 = str(str(year) + "-" + month + "-" + "31")

    df_custom = df.loc[(df['data'] >= data1) & (df['data'] <= data2)]
    df_custom = changeIndex(df_custom)


    fig3 = plt.figure(figsize=(10, 5))
    plt.title("Temperature graph in Celsius by month")  # Fahrenheit temperature graph
    plt.xlabel("Day")
    plt.ylabel("Temp")
    plt.plot(df_custom['temp'], color="blue", linestyle="--", marker="x", linewidth=1, markersize=4)
    plt.axhline(y=np.nanmean(df_custom["temp"]), color="red", linewidth=1)
    plt.axhline(y=np.nanmedian(df_custom["temp"]), color="orange",linestyle="-", linewidth=2)
    plt.legend(["Temp in day", "avg", "median"])
    plt.show()
```

```python
month, year = 8, 2021
custom(df, month, year)
```
